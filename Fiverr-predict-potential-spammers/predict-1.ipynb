```python
import pandas as pd
df = pd.read_csv("train.csv")
df.head()
```

```python
import numpy as np
print("Input data shape:", np.array(df).shape)
print()
print(df.columns)
```

```python
df = df.drop(columns=["user_id"])
X = df[df.columns[1:]]
y = df["label"]

X = np.array(X)
y = np.array([y]).T
print(X.shape, y.shape)
```

```python
val_size = int(X.shape[0]*0.05)

xtrain = X[val_size:]
xval = X[:val_size]

ytrain = y[val_size:]
yval = y[:val_size]

#xtrain = xtrain.reshape((-1,1))
#xval = xval.reshape((-1,1))
#ytrain = ytrain.reshape((-1,1))
#yval = yval.reshape((-1,1))

print(f"xtrain shape =\t\t{xtrain.shape}\nxvalidation shape =\t{xval.shape}")
print()
print(f"ytrain shape =\t\t{ytrain.shape}\nyvalidation shape =\t{yval.shape}")
```

```python
import tensorflow as tf
import keras as k
import keras.layers as l

np.random.seed(0)
model = k.Sequential([
    l.Input(shape=(51,)),
    l.Dense(126, activation="relu"),
    l.Dense(51, activation="softmax"),
    l.Flatten()
])
model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"])
h = model.fit(x=xtrain, y=ytrain, epochs=10, validation_data=(xval, yval,), batch_size=1000)
model.summary()
```
