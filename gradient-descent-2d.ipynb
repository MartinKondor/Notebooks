## Gradient Descent with one variable

We define a function C and it's derivative for x input.


```python
import numpy as np
import matplotlib.pyplot as plt


def C(x):
    return x**2


def dC(x):
    return 2*x
    

x = np.arange(-10, 10, 0.1)

plt.figure(figsize=(7, 5))
plt.grid()
plt.title("Cost funciton and it's derivate")
plt.plot(x, C(x), c="b", linewidth=2)
plt.plot(x, dC(x), c="r", linewidth=2)
plt.legend(["C", "dC/dx"])
plt.show()
```

Define the optimizer function. The global minimum is at (0,0) so we expect the optimizer to find that point.

```python
def plot_grad_change(x, c, grad_xs, grad_ys):
    plt.figure(figsize=(9, 7,))
    plt.title("Gradient change over time")
    plt.grid()
    plt.scatter(grad_xs, grad_ys, c="r", marker="v", alpha=1., s=150)
    plt.scatter([0], [0], c="g", marker="o", alpha=0.7, s=150)
    plt.plot(x, c(x), c="b", alpha=0.75)
    plt.legend(["C", "Gradient Point", "Global minimum"])
    plt.show()


def SGD(x, c, dc, epochs, eta):
    
    # Get a random point from the input range
    np.random.seed(12)
    px = np.random.random() * np.max(x) + np.min(x)
    py = c(px)
    grad_xs, grad_ys = [px], [py]
    
    for epoch in range(epochs):
        px += -eta*dc(px)
        py = c(px)
        grad_xs.append(px)
        grad_ys.append(py)
        

    plot_grad_change(x, c, grad_xs, grad_ys)


epochs = 15
learning_rate = 0.1
SGD(x, C, dC, epochs, learning_rate)
```
